**using "reshape" method to reshaping the no of rows of rows and colums of a array**


```python
import numpy as np
a=np.array([[1,2],[4,5],[7,9]])
print(a.shape)
print(a.reshape(2,3))
print()
print()
print(a.reshape(6,1))
print()
print()
print(a.reshape(1,6))
```

**using "ravel()" method to flatten a arra========>>>>>>>> here ravel and all the above functions don't changes the original array it creates an another array **


```python
import numpy as np
a=np.array([[1,2],[4,5],[7,9]])
print(a.ravel())
print()
print(a)            # here original array will be printed
```

**"min()" method gives the min value element and "max()" method will return the max value element and "sum()" method will sum all elements of the array**

```python
import numpy as np
a=np.array([[1,2],[4,5],[7,9]])
print("The min value element of the given array is:",a.min())
print("The max value element of the given array is:",a.max())
print("The sum of all the eleents is:",a.sum())
```

**axis 0 is used to denote the columns and axis 1 is used to denote the rows         ===========>>>>>>>>>>>>>>>> To get the sum of the elements of a array column-wise we use axis = 0 and for row-wise axis = 1**

```python
import numpy as np
a=np.array([[1,2],[4,5],[7,9]])
print(a)
print("column-wise sum:",a.sum(axis=0))
print()
print("Row-wise sum:",a.sum(axis=1))
```

**using "sqrt()" method to get the square root of all elements of a array**

```python
import numpy as np
a=np.array([[1,2],[4,5],[7,9]])
print(a)
print("Square root of the array is:")
np.sqrt(a)
```

**using "std()" method to get the standard deviation for the given matrix**

```python
import numpy as np
a=np.array([[1,2],[4,5],[7,9]])
np.std(a)
```

**using "array1.dot(array2)" to get the matrix product of two arrays**

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.array([[7,8],[9,10]])
print(a)
print(b)
print("matrix product of matrix a and matrix b is:")
print(a.dot(b))
print("matrix product of b and a is:")
print(b.dot(a))
```
